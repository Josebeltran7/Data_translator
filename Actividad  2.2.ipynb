## `subplots()` Avanzado

`plt.subplots(n,m)` permite especificar una malla de `n` filas por `m` columnas, con un conjunto de ejes en cada mallado. Los conjuntos de ejes se muestran como un array de numpy.

```python
fig, axs = plt.subplots(2,4)
fig.tight_layout()
```

```python
type(axs[0])
```

```python
axs[0,2]
```

Para graficar en uno de los conjuntos de ejes se puede utilizar su índice en el array del mallado y después graficar en este espacio.

```python
fig, axs = plt.subplots(2,4)
axs[0,2].plot(x_sin,y_sin)
fig.tight_layout()
```

#### Tamaño de subplots
El tamaño de los subplots también puede configurarse con la instrucción `figsize` dentro de `plt.subplots()`.

```python
fig, axs = plt.subplots(2,4, figsize=(10,4))

axs[0,2].plot(x_sin,y_sin)
axs[0,2].set_title('cool')

fig.tight_layout()
```

### Comunmente se utiliza `ax` para un solo conjunto de ejes, y se utiliza `axs` para más de un conjunto de ejes.

#### Iteraciones entre conjuntos de ejes en un subplot

Cuando se tiene un gran mallado de conjuntos de ejes, indizar manualmente cada uno de ellos se vuelve complicado, por lo que es mas sencillo utilizar un ciclo.

Solo necesitamos escribir `for ax in axs`, aunque al iterar entre un np.array de `n` por `m` el resultado es una _fila_ a la vez, más no un _elemento_. Es por esto que se tiene que utilizar la instrucción `.flatten()` para transformar el array a una sola dimensión.

```python
axs.flatten().shape
```

```python
fig, axs = plt.subplots(2,4, figsize=(10,4))

for i, ax in enumerate(axs.flatten()):
    ax.scatter(x_sin,y_sin*(i+1))
    ax.set_title('Gráfico número {}'.format(i))

fig.tight_layout()
```

## plt.plot

Existe una manera más veloz de crear gráficos.

```python
plt.scatter(x_data, y_data)
```

And MANY online examples will use this method to plot.

```python
#this is OK

plt.scatter(x_data, y_data)
plt.title('Line')
plt.xlabel('x')
plt.ylabel('y')
```

Sin embargo, este método se vuelve demasiado complicado en el momento en que se añade más que los títulos.

```python
fig = plt.figure()

ax1 = fig.add_subplot(1,2,1)
plt.title('Seno y Seno**2')

fig1 = plt.gca()

ax2 = fig.add_subplot(1,2,2)
plt.title('Línea')

fig2 = plt.gca()

ax1_stuff = ax1.plot(x_sin, y_sin, color='m', alpha=.25)

ax1.plot(x_sin, y_sin**2)

# ax1_stuff[0].set_alpha(.25) # Método para utilizar el parámetro alpha

ax2.plot(x_sin, x_sin)

line = ax1_stuff[0]
line.set_alpha(.55)

fig.suptitle('Dos líneas', fontsize=44, y=1.1)
fig.tight_layout()
```

Por esto se recomienda utilizar `fig, axs = plt.subplots()`

# Escribir funciones para crear gráficos

**Siempre que tenga el instinto de copiar y pegar es señal de que necesitas escribir una función.**

La mejor manera de escribir una función con gráficos es hacer una función que tome un eje y algunos datos y los grafique en ese eje.

```python
def my_plotter(ax, data1, data2, param_dict):
    """
    A helper function to make a graph

    Parameters
    ----------
    ax : Axes
        The axes to draw to

    data1 : array
       The x data

    data2 : array
       The y data

    param_dict : dict
       Dictionary of kwargs to pass to ax.plot

    Returns
    -------
    out : list
        list of artists added
    """
    out = ax.plot(data1, data2, **param_dict)
    return out
```

```python
fig, ax = plt.subplots()
out = my_plotter(ax, x_sin, y_sin, {'linewidth':3, 'linestyle':'-.'})
```
